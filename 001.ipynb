```python
# load env vars
from dotenv import load_dotenv

load_dotenv()
```

```python
# create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# make a request
message = client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = [
        {
            "role": "user",
            "content": "Who were Cerberus's brothers and sisters? Only provide names."
        }
    ]
)
```
